```python

import numpy as np              #1st_step=import libraryies
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv ('Lprice.csv')    #2nd_step=read csv files
                          #3rd_step=Divided into two parts
i = df.iloc [:,0].values           #It is a independent features
d = df.iloc [:,1].values           #It is a dependent features

print(i)
print(d)
df
```

```python
from sklearn.model_selection import train_test_split       #import library for test/split 
#from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split (i, d, test_size = .20)      #divided into for parts

print("records in x_test")
print(x_test)  # 8 records in x_test
print(y_test) 
#print(x_train)
#print(y_train)	


#print("output of x_train before reshape is :")
#print(x_train)
#x_train = x_train.reshape(-1,1)  # reshape(-1,1) convert the 1D data into 2D data
#print(x_train)

#y_train = y_train.reshape(-1,1)


#print("output of x_train after reshape is :")
#print(x_train)
```

```python
#print("again reshaping")
#x_train = x_train.reshape(1,-1)#reshape(1,-1)convert again into original data 
#print(x_train)
x_test = x_test.reshape(-1,1)

from sklearn.linear_model import LinearRegression          #import algorithm for linear regression
lnr = LinearRegression()         #create object for linear regression
lnr.fit (x_train, y_train )             #fit fxn used for traning data

predicted_y = lnr.predict(x_test)   #predict fxn used for testing data
print("predicted y records:")
print(predicted_y)
print(y_test)
#print(x_test)
```

```python
#import math libraryes
#find error rate

from sklearn import metrics
from sklearn.metrics import mean_squared_error
from math import sqrt
print("RMSE is:-")

print('Mean Absolute Error(MAE):', metrics.mean_absolute_error(y_test, predicted_y))
print('Mean Squared Error(MSE):', metrics.mean_squared_error(y_test, predicted_y))
print('Root Mean Squared Error(RMSE):', np.sqrt(metrics.mean_squared_error(y_test, predicted_y)))
```

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test,predicted_y, color = 'green')
plt.title ('Results ')
plt.xlabel('Area of Flats')
plt.ylabel('Price')
plt.show()
```
